```python
import matplotlib.pyplot as plt
import numpy
from functions import get_optimal_nodes, lagrange_interpolation
```

## The function

```python
def f(x):
    return 3 / (x*x + 4*x + 5)
```

```python
def get_index(n):
    
    if (2 <= n <= 4):
        return 0
    
    if (5 <= n <= 7):
        return 1
    
    if (8 <= n <= 15):
        return 2
    
    if (16 <= n <= 21):
        return 3
    
    return 4
```

```python
fig, ax  = plt.subplots(nrows=4, ncols=1, figsize=(16, 15))

left, right = -12, 12

density = 10000

t = numpy.linspace(left, right, density, dtype="float128")

for n in range(2, 22):
    
    x = get_optimal_nodes(n=n, left=left, right=right)
    y = f(x)
    
    X, Y = lagrange_interpolation(x, y, left=left, right=right, density=density)
    
    ax[get_index(n)].plot(X, Y, "--")

    
    
ax[0].plot(t, f(t), color="r")
ax[1].plot(t, f(t), color="r")
ax[2].plot(t, f(t), color="r")  
ax[3].plot(t, f(t), color="r")

ax[0].legend([f"{i - 1}" for i in range(2, 5)] + [""])
ax[1].legend([f"{i - 1}" for i in range(5, 8)] + [""])
ax[2].legend([f"{i - 1}" for i in range(8, 16)] + [""])
ax[3].legend([f"{i - 1}" for i in range(16, 22)] + [""])

#ax[0].axis("equal")

plt.show()
```
