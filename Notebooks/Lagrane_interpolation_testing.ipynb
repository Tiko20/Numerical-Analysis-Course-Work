```python
def sgn(x):
    return numpy.nan_to_num(x / numpy.abs(x), 0)
```

```python
def f(x):
    return numpy.abs(x) ** x
```

```python
left, right, dens = -2, 2, 1000
nodes = 29
```

```python
X = get_optimal_nodes(nodes, left, right)
#X = numpy.linspace(left, right, nodes)
Y = f(X)
```

```python
F = LagrangeInterpolation(X, Y)
```
